```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from selenium import webdriver
import time
schools = {}  
items = []
def get_links():
    index = 0
    base_url = "https://www.switchup.org/" 
    driver = webdriver.Edge()
    for i in range(1,11):
        items = []
        url = f'https://www.switchup.org/coding-bootcamps-reviews?page={i}'
        driver.get(url)
        time.sleep(5)
        soup = BeautifulSoup(driver.page_source, 'html.parser')
        items = soup.find_all("a", attrs = {"class":"bootcamp-list__row--review-count"})
        items = items[3:]
        for i in range(len(items)):
            constructed_url = f"{base_url}{items[i]['href']}"
            schools[index] = constructed_url
            index +=1
    return schools
get_links()
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup

url = "https://www.switchup.org/coding-bootcamps-reviews"
base_url = "https://www.switchup.org/" 
driver = webdriver.Edge()
driver.get(url)

links = []
index = 0

while True:
    soup = BeautifulSoup(driver.page_source, 'html.parser')
    links_on_page = [a['href'] for a in soup.find_all("a", attrs = {"class":"bootcamp-list__row--review-count"})]
    links.extend(links_on_page[3:])
    try:
        next_button = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH, "//button[contains(., 'Next')]")))
        next_button.click() 
    except Exception:
        break 
driver.quit()

for i in range(len(links)):
    constructed_url = f"{base_url}{links[i]}"
    schools[index] = constructed_url
    index +=1
schools
```

```python
schools_main = {}
for i in schools.values():
    r = requests.get(i)
    soup = BeautifulSoup(r.content)
    x = soup.find_all("page-data")[0]["school-id"]
    schools_main[i.split("/")[-1]]=x
```
